# Parse one file

```python
def parse_bloomberg_data(filename, sheet_names=['Income Statement', 'Balance Sheet', 'Cash Flow', 'Reconciliation', 'Ratios']):
    dataframes = []
    for sheet in sheet_names:
        if sheet == 'Ratios':
            dataframes.append(pd.read_excel(filename, sheet_name=sheet, index_col=0).dropna(axis=0, how='all') / 100)
        else:
            dataframes.append(pd.read_excel(filename, sheet_name=sheet, index_col=0).dropna(axis=0, how='all'))
    res = pd.concat(dataframes)
    res = res[~res.index.duplicated(keep='first')].fillna(method='pad').T
    res.to_excel(filename)
```

```python
import os
for filename in os.listdir('report data'):
    path = 'report data/' + filename
    print(path)
    parse_bloomberg_data(path)
```

```python
def parse_dates(filename):
    res = pd.read_excel(filename)
    res.rename(columns={ res.columns[0]: "Date" }, inplace = True)
    res['Date'] = res['Date'].astype(str).str.replace(" ", "")
    res['Date'] = pd.to_datetime(res['Date'])
    res['Date'] = res['Date'] + pd.offsets.QuarterEnd(0)
    res.to_excel(filename)
```

```python
import os
for filename in os.listdir('report data'):
    path = 'report data/' + filename
    print(path)
    parse_dates(path)
```

```python
def parse_market_cap_data(filename='market_cap.xlsx'):
    market_cap = pd.read_excel('market_cap.xlsx')
    day_exclusion = ['Saturday', 'Sunday']
    market_cap = market_cap[~(pd.to_datetime(market_cap['Date']).dt.day_name().isin(day_exclusion))].reset_index(drop=True)
    return market_cap
```

```python
market_cap = parse_market_cap_data()
```

```python
market_cap['Date'] = pd.to_datetime(market_cap['Date'])
```

```python
market_cap = market_cap.dropna(how='all')
```
